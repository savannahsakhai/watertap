```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
#import plotly.express as px
```

```python
# Reverse Osmosis

RO_NaCl = pd.read_csv('outputs_results_RO_NaCl.csv')
RO_NaCl = RO_NaCl.dropna()
RO_NaCl = RO_NaCl.reset_index(drop=True)

RO_Sea = pd.read_csv('outputs_results_RO_seawater.csv')
RO_Sea = RO_Sea.dropna()
RO_Sea = RO_Sea.reset_index(drop=True)

RO_Simple = pd.read_csv('outputs_results_RO_simple.csv')
RO_Simple = RO_Simple.dropna()
RO_Simple = RO_Simple.reset_index(drop=True)
```

```python
fig1 = plt.figure('Membrane Area')
#ax = fig.add_subplot(projection='2d')
plt.plot(RO_NaCl['# Water Recovery'], RO_NaCl['Membrane Area'])
plt.plot(RO_Sea['# Water Recovery'], RO_Sea['Membrane Area'])
plt.plot(RO_Simple['# Water Recovery'], RO_Simple['Membrane Area'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Membrane Area')
```

```python
fig2 = plt.figure('Operating Pressure')
#ax = fig.add_subplot(projection='2d')
plt.plot(RO_NaCl['# Water Recovery'], RO_NaCl['Operating Pressure'])
plt.plot(RO_Sea['# Water Recovery'], RO_Sea['Operating Pressure'])
plt.plot(RO_Simple['# Water Recovery'], RO_Simple['Operating Pressure'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Pressure')
```

```python
fig = plt.figure('LCOW')
#ax = fig.add_subplot(projection='2d')
plt.plot(RO_NaCl['# Water Recovery'], RO_NaCl['LCOW'])
plt.plot(RO_Sea['# Water Recovery'], RO_Sea['LCOW'])
plt.plot(RO_Simple['# Water Recovery'], RO_Simple['LCOW'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('LCOW')
```
