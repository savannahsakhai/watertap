```python
# Original parameter names and grouping labels
parameters = ["HCl cost", "NaOH cost", "Electricity cost", "Water permeability", 
              "Water permeability", "Salt permeability", "Salt permeability",
              "Boron rejection in 1st pass RO", "Membrane cost", "Membrane replacement rate", 
              "Pump capital cost", "Mixer capital cost", "Boron rejection in 2nd pass RO $^{*}$","Feed pH of 1st pass RO"]

# Sample data for LCOW_low and LCOW_high (the reordered values should match your data)
LCOW_low = [s1["LCOW"][0], s2["LCOW"][0], s3["LCOW"][0], s4["LCOW"][0], s5["LCOW"][0], 
            s6["LCOW"][0], s7["LCOW"][0], s8["LCOW"][0], s9["LCOW"][0], s10["LCOW"][0], 
            s11["LCOW"][0], s12["LCOW"][0], s13["LCOW"][0], s14["LCOW"][0]]

LCOW_high = [s1["LCOW"][1], s2["LCOW"][1], s3["LCOW"][1], s4["LCOW"][1], s5["LCOW"][1], 
             s6["LCOW"][1], s7["LCOW"][1], s8["LCOW"][1], s9["LCOW"][1], s10["LCOW"][1], 
             s11["LCOW"][1], s12["LCOW"][1], s13["LCOW"][1], s14["LCOW"][1]]

# New ordered indices based on the new parameter order
indices_order = [13, 7, 12, 8, 9, 10, 11, 0, 1, 2]

# Reorder LCOW_low and LCOW_high based on the new indices
parameters = [parameters[i] for i in indices_order]
LCOW_low = [LCOW_low[i] for i in indices_order]
LCOW_high = [LCOW_high[i] for i in indices_order]


fig = fig_generator.figureGenerator()
fig.init_figure()

# mid_dark= "#07578c"
# mid_light= "#85cced"
econ =  "#b2df8a"
color = [dark,
        mid_dark,
        light,
    econ,econ,econ,econ,
    econ,econ,econ,econ,]

y_pos = np.arange(len(parameters))

width = []
bottom = []
for low, high in zip(LCOW_low, LCOW_high):
    width.append(abs(high - low))
    bottom.append(min(low, high))

# Loop over each parameter to plot individual bars with unique labels
for i, (y, w, b, label,c) in enumerate(zip(y_pos, width, bottom, parameters,color)):
    fig.plot_bar(
        [y],  # single y position
        [w],  # single width
        bottom=[b],  # single bottom position
        width=0.3,
        vertical=False,
        color=c,
        save_label=label,  # individual label for each bar
    )

# Additional plot settings
plt.yticks(y_pos, parameters)
plt.xlim(-13, 13)
plt.xlabel('Change in LCOW (%)')
plt.gca().invert_yaxis()
plt.show()
```
