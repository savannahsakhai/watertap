```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
```

```python
# read in data

data_NaCl = pd.read_csv('outputs_results_NaCl.csv')
data_Sea = pd.read_csv('outputs_results_Sea.csv')
data_Simple = pd.read_csv('outputs_results_simple.csv')
```

```python
%matplotlib ipympl
fig = plt.figure('Enthalpy')
ax = fig.add_subplot(projection='3d')
ax.scatter(data_NaCl['# Temperature'], data_NaCl['Flow'], data_NaCl['Enthalpy'], c='blue')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], data_Sea['Enthalpy'], c='orange')
ax.scatter(data_Simple['# Temperature'], data_Simple['Flow'], data_Simple['Enthalpy'], c='green')

ax.set_xlabel('Temperature, K')
ax.set_ylabel('Salinity')
ax.set_zlabel('Enthalpy, J/kg')
fig.show()
```

```python
%matplotlib ipympl
fig = plt.figure('Vap Pressure')
ax = fig.add_subplot(projection='3d')
ax.scatter(data_NaCl['# Temperature'], data_NaCl['Flow'], data_NaCl['Vap Pressure'], c='blue')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], data_Sea['Vap Pressure'], c='orange')
ax.scatter(data_Simple['# Temperature'], data_Simple['Flow'], data_Simple['Vap Pressure'], c='green')

ax.set_xlabel('Temperature, K')
ax.set_ylabel('Salinity')
ax.set_zlabel('Vap Pressure, Pa')
fig.show()
```

```python
%matplotlib ipympl
fig = plt.figure('Osmotic Pressure')
ax = fig.add_subplot(projection='3d')
ax.scatter(data_NaCl['# Temperature'], data_NaCl['Flow'], data_NaCl['Osmotic Pressure']/10**5, c='blue')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], data_Sea['Osmotic Pressure']/10**5, c='orange')
ax.scatter(data_Simple['# Temperature'], data_Simple['Flow'], data_Simple['Osmotic Pressure']/10**5, c='green')

ax.set_xlabel('Temperature, K')
ax.set_ylabel('Salinity')
ax.set_zlabel('Osmotic Pressure, bar')

fig.show()
```

```python
rel_error_NaCl = (data_NaCl['Osmotic Pressure'] - data_Simple['Osmotic Pressure'])/data_NaCl['Osmotic Pressure']
rel_error_Sea_Simple = (data_Sea['Osmotic Pressure'] - data_Simple['Osmotic Pressure'])/data_Sea['Osmotic Pressure']
rel_error_Sea_NaCl = (data_Sea['Osmotic Pressure'] - data_NaCl['Osmotic Pressure'])/data_Sea['Osmotic Pressure']
#print(abs_error)

%matplotlib ipympl
fig = plt.figure('Osmotic Pressure Error')
ax = fig.add_subplot(projection='3d')
#ax.scatter(data_NaCl['# Temperature'], data_NaCl['Flow'], rel_error_NaCl, c='blue')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], rel_error_Sea_Simple, c='green')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], rel_error_Sea_NaCl, c='blue')
```

```python
%matplotlib ipympl
fig = plt.figure('Osmotic Coeff')
ax = fig.add_subplot(projection='3d')
ax.scatter(data_NaCl['# Temperature'], data_NaCl['Flow'], data_NaCl['Osmotic Coeff'], c='blue')
ax.scatter(data_Sea['# Temperature'], data_Sea['Flow'], data_Sea['Osmotic Coeff'], c='orange')

ax.set_xlabel('Temperature, K')
ax.set_ylabel('Salinity')
ax.set_zlabel('Osmotic Coeff')
fig.show()
```
