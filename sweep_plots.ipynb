```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt

# from analysis_plot_kit.core import fig_generator
```

```python
data = pd.read_csv('sensitivity_full_flowsheet1.csv')
data.dropna()
```

```python
fig = plt.figure()

data1 = data[(data["boron_limit"] == 3.00000e-04)]
plt.plot(data1['# boron_feed']*1000, data1['RO2_Boron_rej'], 'o')
data1 = data[(data["boron_limit"] == 5.00000e-04)]
plt.plot(data1['# boron_feed']*1000, data1['RO2_Boron_rej'], 'o')
data1 = data[(data["boron_limit"] == 1.0000e-03)]
plt.plot(data1['# boron_feed']*1000, data1['RO2_Boron_rej'], 'o')
data1 = data[(data["boron_limit"] == 2.4000e-03)]
plt.plot(data1['# boron_feed']*1000, data1['RO2_Boron_rej'], 'o')
plt.legend(["0.30 mg/L","0.50 mg/L","1.00 mg/L","2.40 mg/L"],title="Boron regulation", loc='lower right')

plt.xlabel("Inlet Boron Concentration, mg/L")
plt.ylabel("Boron Rejection in 2nd Pass")
```
