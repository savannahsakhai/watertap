```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
```

```python
# read in data

data_NaCl = pd.read_csv('outputs_results_NaCl.csv')
data_Sea = pd.read_csv('outputs_results_Sea.csv')
data_Simple = pd.read_csv('outputs_results_simple.csv')

temperatures = np.unique(np.array(data_NaCl['# Temperature']))
massfrac = np.unique(np.array(data_NaCl['Flow']))
```

```python
fig, (ax1, ax2, ax3)= plt.subplots(3, sharey=True, sharex=True)
temp = [0, 5, 10, 15, 20, 25]
for i in temp:
    a = 0 + i*25
    b = 24 + i*25
    ax1.plot(data_NaCl['Flow'][a:b], data_NaCl['Enthalpy'][a:b])
    ax2.plot(data_Sea['Flow'][a:b], data_Sea['Enthalpy'][a:b])
    ax3.plot(data_Simple['Flow'][a:b], data_Simple['Enthalpy'][a:b])

fig.text(0.5, 0.03, 'Mass Fraction of Salt', ha='center', va='center')
fig.text(0.01, 0.5, 'Enthalpy, J/kg', ha='center', va='center', rotation='vertical')

ax1.set_title('NaCl')
ax2.set_title('Sea')
ax3.set_title('Simple')    
plt.show()
```

```python
fig, (ax1, ax2, ax3)= plt.subplots(3, sharey=True, sharex=True)
temp = [0, 5, 10, 15, 17, 20, 22, 25]
for i in temp:
    a = 0 + i*25
    b = 24 + i*25
    ax1.plot(data_NaCl['Flow'][a:b], data_NaCl['Vap Pressure'][a:b])
    ax2.plot(data_Sea['Flow'][a:b], data_Sea['Vap Pressure'][a:b])
    ax3.plot(data_Simple['Flow'][a:b], data_Simple['Vap Pressure'][a:b])

fig.text(0.5, 0.03, 'Mass Fraction of Salt', ha='center', va='center')
fig.text(0.02, 0.5, 'Vapor Pressure, Pa', ha='center', va='center', rotation='vertical')

ax1.set_title('NaCl')
ax2.set_title('Sea')
ax3.set_title('Simple')    
plt.show()
        
```

```python
fig, (ax1, ax2, ax3)= plt.subplots(3, sharey=True, sharex=True)
temp = [1, 5, 10, 15, 20, 25]
osm_coef_simple = np.repeat(1,24)

for i in temp:
    a = 0 + i*25
    b = 24 + i*25
    ax1.plot(data_NaCl['Flow'][a:b], data_NaCl['Osmotic Coeff'][a:b])
    ax2.plot(data_Sea['Flow'][a:b], data_Sea['Osmotic Coeff'][a:b])
       
ax3.plot(data_Simple['Flow'][0:24],osm_coef_simple)

fig.text(0.5, 0.03, 'Mass Fraction of Salt', ha='center', va='center')
fig.text(0.03, 0.5, 'Osmotic Coeff', ha='center', va='center', rotation='vertical')

ax1.set_title('NaCl')
ax2.set_title('Sea')
ax3.set_title('Simple')  
plt.show()
```
