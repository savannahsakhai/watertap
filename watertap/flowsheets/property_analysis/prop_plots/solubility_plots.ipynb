```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
import matplotlib


# from analysis_plot_kit.core import fig_generator
from psPlotKit.data_plotter.fig_generator import figureGenerator
```

```python
data_total = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\prop_data\solubility_data.csv')
```

```python
data = data_total[data_total["pH"]==7]
data = data.reset_index(drop=True)
```

```python
vmin = np.min(data["deltaCalcite"])*10**13
vmax = np.max(data["deltaCalcite"])*10**13

zticks = np.linspace(vmin, vmax, 5)
zticks = np.around(zticks, decimals=5)
figure = figureGenerator()
figure.init_figure()
figure.colorMaps["color_map"] = cm.get_cmap("GnBu_r")
# figure.add_text(-0.75, 8.5, "A")
# figure.set_title("Reaktoro")
figure.plot_map(
    xdata=data["Conc"],
    ydata=data["T"],
    zdata=data["deltaCalcite"]*10**13,
    text=True,
    text_color="black",
    build_map=True,
    vmin=vmin,
    vmax=vmax,
)
figure.add_colorbar(zticks=zticks, zformat=0, zlabel="Calcite solubility mol/kg water")
xticks = ([1,2,3,4,5,6,7])
yticks = ([20,30,40,50,60,70,80,90,100])
figure.set_axis_ticklabels(
    xticklabels=xticks,
    yticklabels=yticks,
    xformat=1,
    yformat=0,
    xlabel="Concentration (g/kg)",
    ylabel="Temperature (C)",
)

figure.save_fig(name="solubility_calcite")
```
