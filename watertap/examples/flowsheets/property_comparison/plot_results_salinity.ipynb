```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt

%matplotlib ipympl

```

```python
# Reverse Osmosis

RO_NaCl = pd.read_csv('sal_outputs_results_RO_NaCl.csv')
RO_NaCl = RO_NaCl.dropna()
RO_NaCl = RO_NaCl.reset_index(drop=True)

RO_Sea = pd.read_csv('sal_outputs_results_RO_seawater.csv')
RO_Sea = RO_Sea.dropna()
RO_Sea = RO_Sea.reset_index(drop=True)

RO_Simple = pd.read_csv('sal_outputs_results_RO_simple.csv')
RO_Simple = RO_Simple.dropna()
RO_Simple = RO_Simple.reset_index(drop=True)
```

```python
fig = plt.figure('Membrane Area')
ax = fig.add_subplot(projection='3d')
ax.scatter(RO_NaCl['# Water Recovery'], RO_NaCl['Inlet Salinity'], RO_NaCl['Membrane Area'])
ax.scatter(RO_Sea['# Water Recovery'], RO_Sea['Inlet Salinity'], RO_Sea['Membrane Area'])
ax.scatter(RO_Simple['# Water Recovery'], RO_Simple['Inlet Salinity'], RO_Simple['Membrane Area'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Inlet Salinity')
plt.title('Membrane Area')
plt.show()
```

```python
fig = plt.figure('Operating Pressure')
ax = fig.add_subplot(projection='3d')
ax.scatter(RO_NaCl['# Water Recovery'],RO_NaCl['Inlet Salinity'], RO_NaCl['Operating Pressure'])
ax.scatter(RO_Sea['# Water Recovery'], RO_Sea['Inlet Salinity'], RO_Sea['Operating Pressure'])
ax.scatter(RO_Simple['# Water Recovery'], RO_Simple['Inlet Salinity'], RO_Simple['Operating Pressure'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Inlet Salinity')
plt.title('Operating Pressure')
plt.show()
```

```python
fig = plt.figure('LCOW')
ax = fig.add_subplot(projection='3d')
ax.scatter(RO_NaCl['# Water Recovery'], RO_NaCl['Inlet Salinity'], RO_NaCl['LCOW'])
ax.scatter(RO_Sea['# Water Recovery'], RO_Sea['Inlet Salinity'],RO_Sea['LCOW'])
ax.scatter(RO_Simple['# Water Recovery'], RO_Simple['Inlet Salinity'], RO_Simple['LCOW'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Inlet Salinity')
plt.title('LCOW')
plt.show()
```

```python
fig = plt.figure('SEC')
ax = fig.add_subplot(projection='3d')
ax.scatter(RO_NaCl['# Water Recovery'], RO_NaCl['Inlet Salinity'], RO_NaCl['SEC'])
ax.scatter(RO_Sea['# Water Recovery'], RO_Sea['Inlet Salinity'], RO_Sea['SEC'])
ax.scatter(RO_Simple['# Water Recovery'], RO_Simple['Inlet Salinity'], RO_Simple['SEC'])
plt.legend(['NaCl','Sea','Simple'])
plt.xlabel('Water Recovery')
plt.ylabel('Inlet Salinity')
plt.title('SEC')
plt.show()
```

```python
# MVC

MVC_NaCl = pd.read_csv('sal_outputs_results_MVC_NaCl.csv')
MVC_NaCl = MVC_NaCl.dropna()
MVC_NaCl = MVC_NaCl.reset_index(drop=True)

MVC_Sea = pd.read_csv('sal_outputs_results_MVC_seawater.csv')
MVC_Sea = MVC_Sea.dropna()
MVC_Sea = MVC_Sea.reset_index(drop=True)

MVC_Simple = pd.read_csv('sal_outputs_results_MVC_simple.csv')
MVC_Simple = MVC_Simple.dropna()
MVC_Simple = MVC_Simple.reset_index(drop=True)
```
