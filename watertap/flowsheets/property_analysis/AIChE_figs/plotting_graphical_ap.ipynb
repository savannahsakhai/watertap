```python
import pandas as pd
import numpy as np
from mpl_toolkits import mplot3d
# from matplotlib import cmsweep
from matplotlib.ticker import LinearLocator
import matplotlib.pyplot as plt
```

```python
data_RO_seawater = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_RO_sea.csv')
data_RO_nacl = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_RO_nacl.csv')
data_RO_simple = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_RO_simple.csv')
data_RO_reatoro = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_RO_reaktoro.csv')
```

```python
data_MVC_seawater = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_MVC_sea.csv')
data_MVC_reaktoro = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_MVC_reaktoro.csv')
data_MVC_nacl = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_MVC_nacl.csv')
data_MVC_simple = pd.read_csv(r'C:\Users\sss0031\Documents\GitHub\watertap\watertap\flowsheets\property_analysis\tech_data\data_MVC_simple.csv')
```

```python
sea_color = "#4A7B9D" 
nacl_color = "#467740"
reaktoro_color = "#0A0200"

plt.rcParams.update({'font.size': 18})
plt.rcParams.update({'font.family': 'Arial'})
plt.rcParams.update({'font.weight': 'bold'})
title_size = 18
```

```python
fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(12, 4.5), sharey=False)

# Add a larger title centered over both plots
# fig.suptitle("Cost Optimization of Desalination Technologies\nmin Levelized Cost of Water (LCOW)", 
#              fontsize=title_size+2, fontweight='bold', y=0.925)
n=4
# Reverse osmosis plot
axs[0].plot(data_RO_reatoro['# Water Recovery']*100, data_RO_reatoro['LCOW'],
            "o", ms=14, mfc='none', markevery=3, linewidth=n, color=reaktoro_color, label="Reaktoro", zorder=3)
axs[0].plot(data_RO_seawater['# Water Recovery']*100, data_RO_seawater['LCOW'], 
            "*", ms=12, markevery=3, linewidth=n, color=sea_color, label="Lienhard", zorder=2)
axs[0].plot(data_RO_nacl['# Water Recovery']*100, data_RO_nacl['LCOW'], 
            "s", ms=12, mfc='none', markevery=3, linewidth=n, color=nacl_color, label="NaCl", zorder=1)

axs[0].set_xlabel("Water recovery (%)", fontsize=title_size, fontweight='bold')
axs[0].set_ylabel("LCOW ($\$$/m$^3$)", fontsize=title_size, fontweight='bold')
axs[0].set_title("Reverse osmosis", fontsize=title_size, pad=5, fontweight='bold')
axs[0].set_yticks(np.arange(0.6, 0.9, 0.1))
axs[0].set_xticks([30,40,50])


# Mechanical vapor compression plot
axs[1].plot(data_MVC_reaktoro['# Water Recovery']*100, data_MVC_reaktoro['LCOW'], 
            "o", ms=14, mfc='none', markevery=1, linewidth=n, color=reaktoro_color, label="Reaktoro", zorder=3)
axs[1].plot(data_MVC_seawater['# Water Recovery']*100, data_MVC_seawater['LCOW'],
            "*", ms=12, markevery=1, linewidth=n, color=sea_color, label="Lienhard", zorder=2)
axs[1].plot(data_MVC_nacl['# Water Recovery']*100, data_MVC_nacl['LCOW'], 
            "s", ms=12, mfc='none', markevery=1, linewidth=n, color=nacl_color, label="NaCl", zorder=1)

axs[1].set_xlabel("Water recovery (%)", fontsize=title_size, fontweight='bold')
axs[1].set_ylabel("LCOW ($\$$/m$^3$)", fontsize=title_size, fontweight='bold')
axs[1].set_title("Mechanical vapor compression", fontsize=title_size, pad=5, fontweight='bold')
axs[1].set_yticks(np.arange(4.3, 4.65, 0.1))
axs[1].set_xticks([40,55,70])

# Combine handles and labels from both subplots
handles0, labels0 = axs[0].get_legend_handles_labels()
handles1, labels1 = axs[1].get_legend_handles_labels()
handles = handles0  # All three lines are the same for both, so just use one

legend = fig.legend(
    handles,
    ['Reaktoro', 'Lienhard', 'NaCl'],
    loc="lower center",
    bbox_to_anchor=(0.5, -0.2),
    frameon=False,
    # title='Seawater property models',
    fontsize=title_size+4,
    ncol=3
)
if legend.get_title() is not None:
    legend.get_title().set_ha('center')
    legend.get_title().set_fontweight('bold')
    legend.get_title().set_fontsize(title_size+2)


plt.tight_layout()
plt.savefig('graphical_abstract', dpi=450)
```
